# Data Prepration

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('flipkart_laptop.csv')
df.head()
```

### Splitting the brand name

```python
df['name'].str.split()[0]
```

```python
df['Brand'] = np.nan
for i in range(df.shape[0]):
    df['Brand'][i] = df['name'][i].split()[0]
    
df.head()
```

```python
df['Brand']
```

### Look for missing Values

```python
df.isnull().sum()
```

### Separating Ratings and No. of Reviews

```python
df.num_ratings[0]
```

```python
df.num_ratings[0].replace('\xa0','').split('&')
```

```python
df['num_reviews'] = np.nan
for i in range(df.shape[0]):
    if(not pd.isnull(df['num_ratings'][i])):
        rr = df['num_ratings'][i].replace('\xa0','').split('&')
        rat_temp = rr[0].split()[0]
        rev_temp = rr[1].split()[0]

        if (',' in rat_temp):
            df['num_ratings'][i] = int(rr[0].split()[0].replace(',',''))
        else:
            df['num_ratings'][i] = int(rr[0].split()[0])

        if (',' in rev_temp):
            df['num_reviews'][i] = int(rr[1].split()[0].replace(',',''))
        else:
            df['num_reviews'][i] = int(rr[1].split()[0])
    
        

df.head()
```

```python
df.dtypes
```

###   inga iruku converting

```python
df.dropna(subset=['num_ratings'], inplace=True)
df['num_ratings'] = df['num_ratings'].astype(int)
```

```python
ram = []
for i in df['ram'].str.split(' '):
    print(i[0])
    ram.append(i[0])
```

```python
df['ram'] = df['ram'].astype('int')
```
